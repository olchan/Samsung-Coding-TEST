주어진 시간 T 안에 **도달 가능한 모든 교차로의 개수**를 구하는 문제로, 매 시점마다 가능한 하나의 경로를 선택해서 최대로 몇 개 교차로를 방문할 수 있는가를 따지는 문제가 아니다.

- 경로1: (0,0) → (0,1) → (0,2)
- 경로2: (0,0) → (1,0) → (2,0)

👉 {(0,0), (0,1), (0,2), (1,0), (2,0)} 로 총 5개가 출력되어야 한다.


```python
from collections import deque

# 북 동 남 서
dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]

# 신호등 별 (허용 진입 방향, 이후 방향)
# 0: 북  1: 동  2: 남  3: 서
signal_info = [
    [],
    [1, 0,1,2], [0, 0,1,3],
    [3, 0,2,3], [2, 1,2,3],
    [1, 0,1], [0, 0,3],
    [3, 2,3], [2, 1,2],
    [1, 1,2], [0, 0,1],
    [3, 0,3], [2, 2,3]
    ]

map_size, target_time = tuple(map(int, input().split()))

map_signal = [[[] for _ in range(map_size)] for _ in range(map_size)]

for row in range(map_size):
    for col in range(map_size):
        append_signal = list(map(int, input().split()))
        map_signal[row][col].extend(append_signal)

cross_set = set()

# row, col, direction, time
queue = deque()
queue.append((0,0,0,0))

while queue:
    present_row, present_col, present_direction, present_time = queue.popleft()

    if present_time > target_time:
        print(len(cross_set))
        break

    cross_set.add((present_row, present_col))


    signal_index = map_signal[present_row][present_col][present_time % 4]
    dir_condition, *next_directory = signal_info[signal_index]
    if dir_condition == present_direction:
        for next_dir in next_directory:
            next_row = present_row + dr[next_dir]
            next_col = present_col + dc[next_dir]
            if 0 <= next_row < map_size and 0 <= next_col < map_size:
                queue.append((next_row, next_col, next_dir, present_time+1))
```

- BFS에서는 visited 처리를 해서 중복 계산을 줄여줘야 한다.
- 일반적인 BFS에서는 상태가 단순해서 한 번 방문한 위치는 다시 가도 결과가 같다. 그런데, **이 문제는 시점마다도 교차로에 갔을 떄 도달 가능한 다음 교차로 위치가 달라서 시점에 따른 중복 방문 여부를 확인해주어야 한다. 추가로, 특정 교차로에 방문하는 것과 그 교차로를 통과하는 것은 달라서, 중복성을 판단할 떄 진입 방향까지 고려를 해주어야 한다는 것이 특징이다.**

- 같은 위치 (r, c)라도 시간(time % 4)에 따라 신호등 상태가 바뀌고 들어온 방향(direction)에 따라 통과 가능 여부가 달라진다.
- (row, col, direction, time % 4) 👉 "위치 + 시간 + 방향이 모두 같아야 같은 상태"

```python
from collections import deque

# 북 동 남 서
dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]

# 신호등 별 (허용 진입 방향, 이후 방향)
# 0: 북  1: 동  2: 남  3: 서
signal_info = [
    [],
    [1, 0,1,2], [0, 0,1,3],
    [3, 0,2,3], [2, 1,2,3],
    [1, 0,1], [0, 0,3],
    [3, 2,3], [2, 1,2],
    [1, 1,2], [0, 0,1],
    [3, 0,3], [2, 2,3]
    ]

map_size, target_time = tuple(map(int, input().split()))
map_signal = [[[] for _ in range(map_size)] for _ in range(map_size)]

for row in range(map_size):
    for col in range(map_size):
        append_signal = list(map(int, input().split()))
        map_signal[row][col].extend(append_signal)

# row, col, direction, time
queue = deque()
queue.append((0,0,0,0))

cross_set = set()
# row, col, 방문할 때의 방향, 방문 시간 % 4
visited_cross_with_time = set()


while queue:
    present_row, present_col, present_direction, present_time = queue.popleft()

    if present_time > target_time:
        continue

    if (present_row, present_col, present_direction, present_time % 4) in visited_cross_with_time:
        continue

    cross_set.add((present_row, present_col))
    visited_cross_with_time.add((present_row, present_col, present_direction, present_time % 4))

    signal_index = map_signal[present_row][present_col][present_time % 4]
    dir_condition, *next_directory = signal_info[signal_index]
    if dir_condition == present_direction:
        for next_dir in next_directory:
            next_row = present_row + dr[next_dir]
            next_col = present_col + dc[next_dir]
            if 0 <= next_row < map_size and 0 <= next_col < map_size:
                queue.append((next_row, next_col, next_dir, present_time+1))

print(len(cross_set))
```
