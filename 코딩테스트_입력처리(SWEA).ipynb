> **map을 하기 전에 list 를 적용해주지 않아도 동작하는 이유**

- map(int, "123") 을 실행하는 순간 숫자로 다 바뀌는 게 아닙니다. "앞으로 데이터가 들어오면 하나씩 꺼내서 int 로 바꿔줄게!"라는 계획만 세워둔 상태(Iterator)가 됩니다.

- list() 나 for 문 같은 함수가 "자, 첫 번째 값 줘봐!"라고 요청할 때 비로소 문자열 "123" 에서 "1" 을 꺼내 1 로 바꿔서 전달합니다.

- 문자열 자체가 이미 한 글자씩 꺼낼 수 있는 구조(Iterable)이기 때문에, map 은 별도의 리스트 변환 없이도 문자열에서 직접 한 글자씩 꺼내 올 수 있는 것입니다.

```python
# 한 줄에 있는 공백 없는 한 자리의 숫자들을 각각 숫자로 list에 저장하기
num_list = list(map(int, input()))
# list(map(int, list(input()))) 와 동일하기는 하나,
num_list
```

> 2차원 arr의 가장자리를 0으로 감싸기

```python
matrix_size = int(input())
size = matrix_size + 2

grid = (
    [[0] * size]
    + [[0] + list(map(int, input())) + [0] for _ in range(matrix_size)]
    + [[0] * size]
)

print(*grid, sep = "\n")
```

> **2차원 배열과 방향 벡터**

① 좌표 체계 이해 (Row, Column)
코딩테스트의 2차원 배열은 수학 좌표계와 조금 다릅니다.

- r (row): 행 (세로 위치, 위아래)
- c (column): 열 (가로 위치, 좌우)

- 보통 grid[r][c] 로 접근합니다.

② 방향 벡터 설정 (dr, dc)
- 상, 하, 좌, 우 이동을 리스트로 미리 정의합니다. (순서는 시계 방향이 편리할 때가 많습니다.)

```python
# 상, 우, 하, 좌 (시계 방향 예시)
dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]

# 현재 위치
r, c = 2, 2

for i in range(4):
    nr = r + dr[i] # Next Row
    nc = c + dc[i] # Next Column

    # 맵 범위를 벗어나지 않는지 확인 (필수!)
    if 0 <= nr < 4 and 0 <= nc < 5:
        print(f"{i}번 방향으로 이동 가능: ({nr}, {nc})")
```
