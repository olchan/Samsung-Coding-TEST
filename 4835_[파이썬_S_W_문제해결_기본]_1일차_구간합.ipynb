```python
test_case = int(input())

for tc in range(test_case):
    N, M = map(int, input().split())
    list_num = list(map(int,input().split()))
    list_min_max = []
    for i in range(N-M+1):
        list_min_max.append( sum(list_num[i:i+M]) )

    print(f"#{tc+1}",max(list_min_max) - min(list_min_max))
```
