> lambda 매개변수들 : 반환값

```python
add = lambda a, b: a + b
print(add(4, 5))

nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)  # [2, 4, 6, 8]

nums = [1, 2, 3, 4, 5]
result = list(filter(lambda x: x % 2 == 0, nums))
print(result)  # [2, 4]

check = lambda x: "even" if x % 2 == 0 else "odd"
print(check(7))  # odd


# 1
arr.sort(key=lambda x: x[1])

# 2
def second(x):
  return x[1]

arr.sort(key=second)
```

# **filter(조건함수, 반복가능객체)**

- 결과는 👉 filter 객체 (이터레이터)  -> 그래서 보통 list()로 감싼다.

- “뭔가를 ‘변형’하면서 만들 때는 map + lambda,
이미 만들어진 것 중에서 ‘조건에 맞는 것만’ 남길 때는 filter + lambda”

```python
s = "ABCDefgHI"
result = list(filter(lambda c: c.isupper(), s))
print(result)


arr = [1, '2', 3, 'a', 4]
result = list(map(lambda x: x * x, filter(lambda x: isinstance(x, int), arr)))
print(result)
```

# **“가변형 인자”**

👉 인자의 개수에 제한이 없는 함수


| 문법         | 받는 것       | 형태        |
| ---------- | ---------- | --------- |
| `*args`    | **위치 인자**  | 튜플 `()`   |
| `**kwargs` | **키워드 인자** | 딕셔너리 `{}` |

> **함수 정의할 때 순서가 강제됨**

> def func(a, b, *args, **kwargs):


```python
def f(*args):
    print(args)

"""
f(1)
f(1, 2)
f(1, 2, 3, 4)
"""

def g(**kwargs):
    print(kwargs)

"""
g(a=1, b=2, c=3)

# 출력
{'a': 1, 'b': 2, 'c': 3}
"""

def h(*args, **kwargs):
    print(args)
    print(kwargs)
"""
h(1, 2, x=10, y=20)
(1, 2) # args
{'x': 10, 'y': 20} # kwargs
"""
```

> ###  **all & any**
- 리스트, 튜플과 같은 반복 가능한 객체(Iterable) 내 요소들이 참(True)인지 거짓(False)인지 판단할 때 사용하는 매우 유용한 내장 함수

```python
# all() 함수
# 인자로 받은 모든 요소가 True여야만 True를 반환합니다.
# 하나라도 False가 있으면 결과는 False

# 예시 1
print(all([True, True, True]))   # True
print(all([True, False, True]))  # False (하나가 False)

# 예시 2: 리스트 내포와 함께 사용
numbers = [2, 4, 6, 8]
is_all_even = all([n % 2 == 0 for n in numbers])
print(is_all_even)  # True (모두 짝수이므로)
```

```python
# any 함수
# 인자로 받은 요소 중 단 하나라도 True가 있으면 True를 반환합니다.
# 모든 요소가 False여야만 False를 반환합니다.

# 예시 1
print(any([False, False, True]))  # True (하나가 True)
print(any([False, False, False])) # False (모두 False)

# 예시 2: 조건 검사
scores = [50, 40, 95, 30]
has_pass = any(s >= 90 for s in scores)
print(has_pass)  # True (95점이 있으므로)

# 실전 활용 팁 : 0, 빈 문자열(""), None, 빈 리스트([]) 등은 False로 간주
# 이를 활용해 데이터의 유무를 쉽게 판단 가능

data = [1, 2, 0, 4] # 0은 False로 간주됨

print(all(data)) # False (0이 포함되어 있어서)
print(any(data)) # True (0 이외의 숫자들이 있어서)
```

```python
# if 연습문제 5
# ord : 문자(character) → 문자의 유니코드(Unicode) == ASCII
# chr : 문자의 유니코드(Unicode) == ASCII → 문자(character)

inp = input()

if (65 <= ord(inp)):
  if(ord(inp) <= 96):
    print(f"{inp}(ASCII: {ord(inp)}) => {chr(ord(inp) + 32)}(ASCII: {ord(inp) + 32})")
  elif (ord(inp) >= 97) and (ord(inp) <= 122):
    print(f"{inp}(ASCII: {ord(inp)}) => {chr(ord(inp) - 32)}(ASCII: {ord(inp) - 32})")

else:
  print(f"{inp}(ASCII: {ord(inp)})")
```

```python
# 연습문제 - 반복 13
# 10진수 -> 2진수

num = int(input())
numm = num
maxx = 0

while num > 1:
    num = num // 2
    maxx += 1

listt = [0] * (maxx+1)

for i in range(maxx,-1,-1):
    if(numm >= 2**(i)):
         listt[maxx - i] += 1
         numm -= 2**(i)

print(*listt, sep = '')
```

```python
# 9. 내장함수 - 연습문제 5
"""
"ADCBBBBCABBCBDACBDCAACDDDCAABABDBCBCBDBDBDDABBAAAAAAADADBDBCBDABADCADC"와
같은 문자열이 주어지고, A는 4점, B는 3점, C는 2점, D는 1점이라고 할 때 문자열에 사용된
알파벳 점수의 총합을 map 함수와 람다식을 이용해 구하십시오.
"""

score_dict = {'A' : 4, 'B' : 3, 'C' : 2, 'D' : 1}

inpt = input()
total_score = sum(map(lambda x : score_dict[x],inpt))
print(total_score)
```

```python
# 9. 내장함수 - 연습문제 11

"""
다음의 결과와 같이 'abcdef' 문자열의 각각의 문자를 키로 하고 0~5 사이의 정수를
값으로 하는 딕셔너리 객체를 생성하고, 이 딕셔너리 객체의 키와 값 정보를 출력하는
프로그램을 작성하십시오.
"""

string = "abcdef"
result_dict = {ch : idx for idx, ch in enumerate(string)}

for k, v in result_dict.items():
    print(f"{k}: {v}")
```

```python
# 자료구조 -리스트, 튜플 19
"""
단어를 콤마(,)로 구분해 입력하면 그 단어들을 사전순으로 정렬해 출력하는 프로그램을 작성하시시오.
"""

word_lst = list(input().split(", "))
lst_sort = []

for i in range(len(word_lst)):
    min_word = min(word_lst)
    word_lst.remove(min_word)
    lst_sort.append(min_word)

print(*lst_sort, sep = ', ')
```

```python
# 자료구조 -리스트, 튜플 25
"""
리스트 내포 기능을 이용해 [12, 24, 35, 70, 88, 120, 155]에서
첫번째, 다섯번째, 여섯번째 항목을 제거한 후 리스트를 출력하는 프로그램을 작성하십시오.
"""
lst = [12, 24, 35, 70, 88, 120, 155]
print([x for i, x in enumerate(lst) if i not in (0, 4, 5)])
```
