```python
def solution(s):
    # 최소값을 i에 따라 갱신
    answer = 1001
    start = s
    for i in range(1, len(s) + 1):
        result = []
        s = start
        while s:
            if len(s) > 2 * i:
                # case 1 : 연속 i개와 다음 i개가 다른 경우
                if s[:i] != s[i:2 * i]:
                    result.append(s[:i])
                    s = s[i:]
                # case 2 : 연속 i개와 다음 i개가 같은 경우
                else:
                    repeat_num = 2
                    word = s[:i]
                    s = s[2 * i:]
                    result.append(repeat_num)
                    result.append(word)
                    while len(s) >= i:
                        if s[:i] == word:
                            result[-2] += 1
                            if len(s) > i:ㅇ
                                s = s[i:]
                            else:
                                s = ""
                        else:
                            break


            elif (len(s) == 2 * i) & (s[:i] == s[i:2 * i]):
                result.append(2)
                result.append(s[:i])
                s = ""

            else:
                result.append(s)
                s = ""

        final_result = ""
        result = list(map(str, result))
        for i in result:
            final_result += i
        answer = min(len(final_result), answer)

    return answer
```

### 2020 KAKAO BLIND RECRUITMENT 자물쇠와 열쇠

> https://school.programmers.co.kr/learn/courses/30/lessons/60059

```python
# 2차원 배열의 시계방향 90도 회전
def rotate_90(matrix_2d):
    size = len(matrix_2d)
    new_key = [[0] * size for _ in range(size)]

    for row in range(size):
        for col in range(size):
            new_key[col][size - 1 - row] = matrix_2d[row][col]

    return new_key

def solution(key, lock):
    key_90 = rotate_90(key)
    key_180 = rotate_90(key_90)
    key_270 = rotate_90(key_180)
    answer = False

    m = len(key)
    n = len(lock)

    for case in [key, key_90, key_180, key_270]:
        # NxN 안의 모든 원소가 1이면, answer = True
        # nxn 크기의 lock 주변에 mxm  크기의 key가 모든 범위를 순회하도록 padding
        # 가장 초기에는 mxm 크기 key의 오른쪽 아래 1칸이 lock과 겹치는 지점에서 시작
        padding_map = [[0] * (n + 2*m - 2) for _ in range(n + 2*m - 2)]
        for row in range(m-1, m + n - 1):
            for col in range(m-1, m + n - 1):
                padding_map[row][col] = lock[row - m + 1][col - m +1]

        for start_add_row in range(n+m-1):
            for start_add_col in range(n+m-1):
                test_map = [i[:] for i in padding_map]
                for row in range(m):
                    for col in range(m):
                        test_map[start_add_row + row][start_add_col + col] += case[row][col]

                check = True
                for i in range(m-1, n+m-1):
                    for j in range(m-1, n+m-1):
                        if test_map[i][j] != 1:
                            check = False
                            break
                    if check == False:
                        break

                if check == True:
                    answer = True
                    return answer

    return answer
```

### 삼성전자 SW 역량테스트 3190번 뱀

```python
map_size = int(input())
num_apple = int(input())

# 벽을 포함한 map_size+2 x map_size+2 크기 지도 만들기
wall_map = [[2] * (map_size+2)]
for _ in range(map_size):
    add_row = [2] + [0] * map_size + [2]
    wall_map.append(add_row)
wall_map.append([2] * (map_size+2))

# 사과 위치 map에 입력하기
for _ in range(num_apple):
    row, col = map(int, input().split())
    wall_map[row][col] = 1

# 벽이 있는 곳 or 뱀의 몸이 있어서 닿으면 안 되는 곳 : 2
# 사과가 있어서 뱀의 몸이 길어지는 곳 : 1
# 아무것도 아닌 곳 : 0

# 방향 변환 정보 받기
num_direction_change = int(input())

# 상, 우, 하, 좌
direction_operation_row = [-1, 0, 1, 0]
direction_operation_col = [0, 1, 0, -1]
now_index = 1 # 우가 default

from collections import deque
direction_change = deque([])
for _ in range(num_direction_change):
    time, direction = input().split()
    direction_change.append((int(time)+1, direction))


# 뱀의 위치를 저장하는 heap -> 사과를 만나지 않으면서 이동하는 경우
# 처음 추가된 위치가 빠지도록
snake = deque([(1, 1)])
now_row = 1
now_col = 1
time = 0
# 가장 가까운 방향 전환
if direction_change:
    change_time, direction = direction_change.popleft()

######## START ########
while True:
    time += 1
    # 현재 이동할 방향 정보 update
    if time < change_time:
        pass
    elif change_time == time:
        if direction == 'D':
            now_index = (now_index + 1) % 4
        elif direction == 'L':
            now_index = (now_index - 1) % 4
        if direction_change:
            change_time, direction = direction_change.popleft()

    now_row += direction_operation_row[now_index]
    now_col += direction_operation_col[now_index]

    # 벽이 있다면?
    if wall_map[now_row][now_col] == 2:
        print(time)
        break
    # 사과가 있다면? heap에 뱀 위치만 추가
    elif wall_map[now_row][now_col] == 1:
        snake.append((now_row, now_col))
        wall_map[now_row][now_col] += 1
    # 사과가 없다면? heapq에서 뱀 위치 빼고 그 때의 map에 -2 새로운 위치 추가 후 +2
    elif wall_map[now_row][now_col] == 0:
        del_row, del_col = snake.popleft()
        wall_map[del_row][del_col] = 0
        snake.append((now_row, now_col))
        wall_map[now_row][now_col] = 2
```

### 2020 KAKAO BLIND RECRUITMENT 기둥과 보 설치

https://school.programmers.co.kr/learn/courses/30/lessons/60061?language=python3

> 2차원 List로 벽을 구현해서 각 교차점에 있는 보와 벽의 존재 여부를 판단하려하면, 예외 처리가 너무 많다는 것을 알 수 있다. 코드를 작성하면서, 예외 처리가 너무 많다 싶을 때는 해당 방법 외에는 풀 수 있는 방법이 없는지 확인을 해보자.

❌ 2차원 배열로 벽을 그리는 방식

👉 구조물 간 관계를 확인할 때 불필요한 탐색과 조건 분기가 늘어나 코드가 복잡해지고 실수 확률이 커진다.

⭕ (x, y, 구조물 종류)를 원소로 가지는 집합(set)으로 관리

```python
def install_object(width, height, obj, install_set):
    can_install = False
    # obj가 기둥인 경우
    if obj == 0:
        # 높이가 0, 즉 바닥인 경우
        if height == 0:
            can_install = True
        # 왼쪽 아래 보가 존재하는 경우
        elif (width-1, height, 1) in install_set:
            can_install = True
        # 오른쪽 아래 보가 존재하는 경우
        elif (width, height, 1) in install_set:
            can_install = True
        # 밑에 다른 기둥이 있는 경우
        elif (width, height-1, 0) in install_set:
            can_install = True

    # obj가 보인 경우
    elif obj == 1:
        # 왼쪽에 기둥이 있는 경우
        if (width, height-1, 0) in install_set:
            can_install = True
        # 오른쪽에 기둥이 있는 경우
        elif (width+1, height-1, 0) in install_set:
            can_install = True
        # 왼쪽 오른쪽에 보가 있는 경우
        elif ((width-1, height, 1) in install_set) and ((width+1, height, 1) in install_set):
             can_install = True
    return can_install

def check_can_delete(width, height, obj, install_set):
    can_delete = False
    temporary_list = []
    # 임시로 지워보기
    temporary_list.append((width, height, obj))
    install_set.remove((width, height, obj))

    # 다시 설치해보고 되면, 지워도 된다는 것
    if install_object(width, height, obj,install_set):
        can_delete = True

    # 임시로 지웠던 거 다시 install_set에 추가
    add_object = temporary_list.pop()
    install_set.add(add_object)
    return can_delete

def solution(n, build_frame):
    # n은 set에 넣기 위한 조건으로 사용
    # 설치된 보와 기둥에 대한 정보를 담는 set
    install_set = set()
    for instruct in build_frame:
        width, height, obj, install = instruct
        # case : install = 1 설치인 경우
        if install == 1:
            if install_object(width, height, obj, install_set):
                install_set.add((width, height, obj))

        # case : install = 0 삭제인 경우
        elif install == 0:
            # 현재 지우려는 obj의 정보를 임시 저장 후 install_set에서 제거
            # 최종 제거 여부 : delete_ok
            temporary_list = [(width, height, obj)]

            if (width, height, obj) in install_set:
                install_set.remove((width, height, obj))

            delete_ok = True

            # case: 지우려는 obj가 기둥인 경우
            if obj == 0:
                # case1: 그 기둥에 의해 세워진 왼쪽 보
                if (width-1, height+1, 1) in install_set:
                    if not check_can_delete(width-1, height+1, 1, install_set):
                        delete_ok = False

                # case2: 그 기둥에 의해 세워진 오른쪽 보
                elif (width, height+1, 1) in install_set:
                    if not check_can_delete(width, height+1, 1, install_set):
                        delete_ok = False
                # case3: 그 기둥 위에 세워진 기둥
                elif (width, height+1, 0) in install_set:
                    if not check_can_delete(width, height+1, 0, install_set):
                        delete_ok = False

            # case: 지우려는 obj가 보인 경우
            elif obj == 1:
                # case1: 양쪽 보가 모두 있는 경우
                if ((width-1, height, 1) in install_set)  and  ((width+1, height, 1) in install_set):
                    if (not check_can_delete(width-1, height, 1, install_set)) or (not check_can_delete(width+1, height, 1, install_set)):
                        delete_ok = False
                # case2: 왼쪽 아래 기둥이 있는 경우
                if (width, height-1, 0) in install_set:
                    if not check_can_delete(width, height-1, 0, install_set):
                        delete_ok = False

                # case3: 오른쪽 아래 기둥이 있는 경우
                if (width+1, height-1, 0) in install_set:
                    if not check_can_delete(width, height-1, 0, install_set):
                        delete_ok = False

            if delete_ok == False:
                add_object = temporary_list.pop()
                install_set.add(add_object)

    total_obj = list(install_set)
    total_obj.sort()

    return total_obj
```

```python
def install_object(width, height, obj, install_set):
    can_install = False

    # obj가 기둥인 경우
    if obj == 0:
        # 높이가 0, 즉 바닥인 경우
        if height == 0:
            can_install = True
        # 왼쪽 아래 보가 존재하는 경우
        elif (width-1, height, 1) in install_set:
            can_install = True
        # 오른쪽 아래 보가 존재하는 경우
        elif (width, height, 1) in install_set:
            can_install = True
        # 밑에 다른 기둥이 있는 경우
        elif (width, height-1, 0) in install_set:
            can_install = True

    # obj가 보인 경우
    elif obj == 1:
        # 왼쪽에 기둥이 있는 경우
        if (width, height-1, 0) in install_set:
            can_install = True
        # 오른쪽에 기둥이 있는 경우
        elif (width+1, height-1, 0) in install_set:
            can_install = True
        # 왼쪽 오른쪽에 보가 있는 경우
        elif ((width-1, height, 1) in install_set) and ((width+1, height, 1) in install_set):
             can_install = True
    return can_install

def check_can_delete(width, height, obj, install_set):
    can_delete = False
    temporary_list = []
    # 임시로 지워보기
    temporary_list.append((width, height, obj))
    install_set.remove((width, height, obj))

    # 다시 설치해보고 되면, 지워도 된다는 것
    if install_object(width, height, obj,install_set):
        can_delete = True

    # 임시로 지웠던 거 다시 install_set에 추가
    add_object = temporary_list.pop()
    install_set.add(add_object)
    return can_delete

def solution(n, build_frame):
    # n은 set에 넣기 위한 조건으로 사용
    # 설치된 보와 기둥에 대한 정보를 담는 set
    install_set = set()
    for instruct in build_frame:
        width, height, obj, install = instruct
        # case : install = 1 설치인 경우
        if install == 1:
            if install_object(width, height, obj, install_set):
                install_set.add((width, height, obj))

        # case : install = 0 삭제인 경우
        elif install == 0:
            # 현재 지우려는 obj의 정보를 임시 저장 후 install_set에서 제거
            # 최종 제거 여부 : delete_ok
            temporary_list = [(width, height, obj)]

            if (width, height, obj) in install_set:
                install_set.remove((width, height, obj))

            delete_ok = True

            # 그냥 전체 구조물을 검사하는 것이 더 정확하게 검사 가능하다.
            for weight, height, obj in install_set:
                if not check_can_delete (weight, height, obj, install_set):
                    delete_ok = False

            if delete_ok == False:
                add_object = temporary_list.pop()
                install_set.add(add_object)

    total_obj = list(install_set)
    total_obj.sort()

    return total_obj
```

> ## 백준 15686 치킨배달
https://www.acmicpc.net/problem/15686

> 초기 코드

```python
"""
5 1
1 2 0 0 0
1 2 0 0 0
1 2 0 0 0
1 2 0 0 0
1 2 0 0 0
"""
# 초기 코드는 위의 case에서 선택할 치킨집을 분별하지 못함.
```

```python
map_size, max_num_chicken_house = map(int, input().split())
# 전체 치킨집의 row, col 정보
set_chicken_house = set()
# 전체 집의 row, col 정보
set_house = set()
# 선택된 치킨집의 row, col 정보
selected_chicken_house = set()
# 치킨 거리가 결정된 집의 row, col 정보
determined_house = set()

# 정보 받기
for row in range(1, 1 + map_size):
    row_info = (list(map(int, input().split())))
    for col in range(1, 1 + map_size):
        # 집이면?
        if row_info[col - 1] == 1:
            set_house.add((row, col))
        # 치킨집이면?
        if row_info[col - 1] == 2:
            set_chicken_house.add((row, col))

# (dist, house_row, house_col, chicken_row, chicken_col)을 저장하는 heap
import heapq
dist_info = []
for house in set_house:
    for chicken_house in set_chicken_house:
        # 각 집-치킨집 사이의 치킨 거리 계산
        h_row, h_col = house
        c_row, c_col = chicken_house
        distance = abs(h_row - c_row) + abs(h_col - c_col)
        heapq.heappush(dist_info, (distance, h_row, h_col, c_row, c_col))

# heapq에서 하나씩 빼기
minimum_distance = 0
while dist_info:
    distance, h_row, h_col, c_row, c_col = heapq.heappop(dist_info)
    if (h_row, h_col) not in determined_house:
        determined_house.add((h_row, h_col))
        minimum_distance += distance
        selected_chicken_house.add((c_row, c_col))

    # 종료 조건
    if len(determined_house) == len(set_house):
        print(minimum_distance)
        break

    elif len(selected_chicken_house) == max_num_chicken_house:
        to_add_dist = set_house - determined_house
        while dist_info:
            distance, h_row, h_col, c_row, c_col = heapq.heappop(dist_info)
            if ((h_row, h_col) in to_add_dist) and ((c_row, c_col) in selected_chicken_house):
                to_add_dist.remove((h_row, h_col))
                minimum_distance += distance
                if len(to_add_dist) == 0:
                    print(minimum_distance)
                    break
        break
```

> 개선 코드

최대 M개 고르는 것이 문제 조건이었지만, 치킨집 중 M개를 고르는 조합 중 최소값을 구하면 된다.

👉 치킨집을 하나 더 고른다고 해서 각 집은 “더 가까운 치킨집이 생길 수도 있는 것이지, 기존보다 거리가 더 멀어지지는 않기 때문”


```python
from itertools import combinations

# 도시 크기, 최대 치킨집 개수
city_size, max_chicken_house_num = map(int, input().split())
# 집 정보, 치킨 집 정보 입력 받기
house_list = []
chicken_house_list = []
for row in range(1, city_size + 1):
    information = list(map(int, input().split()))
    for col in range(1, city_size + 1):
        # 집이라면?
        if information[col-1] == 1:
            house_list.append((row,col))
        # 치킨집이라면?
        elif information[col-1] == 2:
            chicken_house_list.append((row, col))

# 치킨 거리를 저장하는 치킨집 x 집 size의 2차원 list
house_num = len(house_list)
chicken_house_num = len(chicken_house_list)
chicken_distance = [[0] * (house_num) for _ in range(chicken_house_num)]

chicken_house_index = 0
for chicken_house in chicken_house_list:
    house_index = 0
    for house in house_list:
        row_chicken, col_chicken = chicken_house
        row_house, col_house = house
        distance = abs(row_chicken - row_house) + abs(col_chicken - col_house)
        chicken_distance[chicken_house_index][house_index] = distance
        house_index += 1
    chicken_house_index += 1

# M개의 치킨집으로 구성된 (치킨집 개수 C M)의 combination
all_case_of_M_ChickenHouse = list(combinations([i for i in range(chicken_house_num)], max_chicken_house_num))

# 최소 치킨 거리 갱신
minimum_distance = 1e9

for case in all_case_of_M_ChickenHouse:
    distance = 0
    calculate_matrix = [2*city_size + 1] * house_num

    for row in case:
        for col in range(house_num):
            calculate_matrix[col] = min(calculate_matrix[col], chicken_distance[row][col])

    minimum_distance = min(minimum_distance, sum(calculate_matrix))

print(minimum_distance)
```

## 2020 KAKAO BLIND RECRUITMENT 외벽 점검

https://school.programmers.co.kr/learn/courses/30/lessons/60062

> KEY IDEA 1
- 원형 → 선형으로 바꿔서 문제를 해결
- ex) [1,5,6,10] → [1,5,6,10,13,17,18,22]

> KEY IDEA 2
- 어디서 시작하느냐에 따라 필요한 친구 수 달라지기에, 모든 weak[i]를 시작점으로 시도

> KEY IDEA 3
- 친구 순서를 어떻게 고려하느냐에 따라서도

```python
from itertools import permutations

def solution(n, weak, dist):
    num_of_weak = len(weak)
    weak.extend([w + n for w in weak]) # 원형 -> 선형

    minimum_friend = float('inf')
    # 각각의 시작점에 대해 순회
    for start_idx in range(num_of_weak):
        # 순열을 통해 친구 순서 정하기
        for friends_order in permutations(dist):

            count = 1
            # 시작 지점 + 첫번째 친구의 거리 사용
            position = weak[start_idx] + friends_order[count - 1]

            # position의 범위를 뛰어 넘는 weak의 index가 다음 시작점이 됨
            for index in range(start_idx, start_idx + num_of_weak):
                if weak[index] > position:
                    count += 1
                    if count > len(dist):
                        break
                    position = weak[index] + friends_order[count - 1]

            minimum_friend = min(minimum_friend, count)

    return -1 if minimum_friend > len(dist) else minimum_friend
```
