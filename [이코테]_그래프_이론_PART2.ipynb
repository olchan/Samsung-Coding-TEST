## 서로소 집합 자료구조 : 서로소 부분 집합들로 나누어진 원소들의 데이터를 처리하기 위한 자료구조

- **union-find :“두 원소가 같은 집합인지 빠르게 판단”하기 위해 사용**

- union : 각각의 2개의 원소가 포함된 집합(트리)을 하나의 집합(트리)으로 합치는 연산 -> 집합의 root 노드가 더 작은 집합의 root 노드가 다른 집합의 root 노드를 부모 노드로 갖게 함

- find : 특정한 원소가 속한 집합이 어떤 집합인지 알려주는 연산 -> x가 속한 트리의 root(대표 노드)를 찾는 함수

> 서로소 집합 자료구조는 트리 자료구조를 이용하여 집합을 표현 (이때의 tree는 부모를 가리키는 트리)

> 사용 예시

- 사이클 판별 : if find(a) == find(b):  → a와 b 사이의 edge가 있을 떄 a와 b의 root 노드가 같다면, 사이클이 존재한다는 것.
- 크루스칼 알고리즘
- 연결 여부 판별을 빠르게 하는 경우, BFS 보다 빠르다.

```python
# STEP0. 노드의 개수 크기의 list를 생성. 각 노드의 부모 노드에 대한 정보가 담겨있음 + 초기에는 각 노드는 자기 자신이 부모 노드
# STEP1. 두 tree를 union 하기 위해서는, 두 트리의 root 노드를 비교해서 더 값이 큰 루트노드가 더 작은 루트 노드를 가리키게 한다.
"""
이 과정에서 루트 노드를 알아내기 위해서는, 트리의 각 원소에서 부모 노드를 계속해서 확인하며 거슬러 올라가야한다는 것이다.
-> 서로소 집합 알고리즘으로 루트를 찾기 위해서는 재귀적으로 부모를 거슬러 올라가야한다.
"""
```

```python
# node 개수 & 간선 개수 입력 받기
node, edge = map(int, input().split())
parent = [0] * (node + 1)

for i in range(1, node + 1):
    parent[i] = i

# 특정 원소과 속한 집합(의 root node) 찾기
""" root 노드를 찾기 위해서는 최대 O(V)의 시간이 소요됨 + M개의 find/union 연산 -> O(VM) """
def find_parent(parent, x):
  if parent[x] != x: # 재귀 함수의 break point
    return find_parent(parent, parent[x])

  return x

############# 개선된 find 함수 ################
""" 재귀 함수를 통해 parent[x]를 x의 root 노드로 update하여 경로 압축 """
def find_parent(parent, x):
  if parent[x] != x:
    parent[x] = find_parent(parent, parent[x])
  return parent[x]

# 두 원소가 속한 집합을 합치기
def union_parent(parent, a, b):
  a = find_parent(parent, a)
  b = find_parent(parent, b)
  if a < b:
    parent[b] = a
  else:
    parent[a] = b

# union 연산 수행
for i in range(edge):
  a, b = map(int, input().split())
  union_parent(parent, a, b)

# 각 원소가 속한 집합 출력
print("각 원소가 속한 집합: ", end = "")
for i in range(1, node+1):
  print(find_parent(parent, i), end = " ")

print()

print("부모 테이블: ", end = "")
for i in range(1,node+1):
  print(parent[i], end = " ")
```

### **서로소 집합을 활용한 CYCLE 판별**
- **무방향 그래프 내에서의 사이클 판별** -> 서로소 집합을 사용하면 해결 가능
- 방향 그래프에서의 사이클 판별 -> DFS

```python
"""간선에 방향성이 없는 무방향 그래프에서의 cycle 판별"""
# "각 간선"을 확인하며 두 node의 root 노드 확인
# case 1. root 노드가 서로 다르다면, 두 노드에 대해 union 연산 수행
# case 2. 어떤 간선을 구성하는 두 노드의 root 노드가 서로 같다면, cycle이 존재하는 것

node, edge = map(int, input().split())
parent = [0] * (node+1)
cycle = False

for i in range(1, node+1):
    parent[i] = i

def find_parent(parent, x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union_parent(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

for i in range(edge):
    a, b = map(int, input().split())
    if find_parent(parent, a) == find_parent(parent, b):
        cycle = True
        break
    else:
        union_parent(parent, a, b)

if cycle:
    print("cycle이 발생했습니다.")
else:
    print("cycle이 발생하지 않았습니다.")
```

> ### 신장 트리 Spanning Tree : 하나의 graph에서 모든 노드를 포함하면서 cycle이 존재하지 않는 부분 graph



## 크루스칼 알고리즘 : 모든 노드를 최소한의 비용으로 연결하는 Minimum Spanning Tree를 만드는 알고리즘

👉 **크루스칼 알고리즘은 그리디 알고리즘이면서, 항상 최적해를 보장한다.**

1. 모든 간선을 비용에 따라 오름차순으로 정렬
2. 간선을 하나씩 확인하면서 현재의 간선이 cycle을 발생시키는지 확인
- cycle 발생하면, minimum spanning tree에 포함 x
- cycle 발생 안 하면, minimum spanning tree에 포함
3. 모든 간선에 대하여 반복

> 이떄, 최종적으로 "**신장 트리에 포함되는 간선의 개수는 node-1개**"라는 특징이 있다.

>> 시간 복잡도 : 간선의 개수가 E개일 때, O(ElogE)

| 연산            | 시간복잡도      |
| ------------- | ---------- |
| 정렬            | O(E log E) |
| find/union 반복 | O(E)       |


```python
# MST(minimum spanning tree)를 구성하는 비용 알아내기

def find_parent(parent, x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union_parent(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

nodes, edges = map(int, input().split())
parent = [0]*(nodes+1)

# 모든 간선을 담을 list & MST의 최종 비용
edge_list = []
result = 0

for i in range(1, nodes+1):
    parent[i] = i

for _ in range(edges):
    a, b, cost = map(int, input().split())
    edge_list.append((cost, a, b))

edge_list.sort() # 정렬의 시간 복잡도 : ElogE

for edge in edge_list:
    cost, a, b = edge
    # edge를 구성하는 두 노드의 root node가 다른, 즉 아직은 연결되지 않아서
    # 두 노드가 서로 다른 그룹에 있다면, mst에 해당 노드 추가
    if find_parent(parent, a) != find_parent(parent, b):
        union_parent(parent, a, b)
        result += cost

print(result)
```

## 위상 정렬 알고리즘 Topology Sort

- 방향 그래프의 모든 노드를 그래프의 "방향성"에 거스르지 않도록 순서대로 나열하는 것

> **진입 차수 : 특정한 노드로 들어오는 간선의 개수**

1. 진입 차수가 0인 노드를 queue에 넣는다.
2. queue가 빌 때까지 다음의 과정을 반복한다.
- 큐에서 원소를 꺼내 해당 노드에서 출발하는 간선을 그래프에서 제거한다.
- 새롭게 진입차수가 0이 된 노드를 queue에 넣는다.

👉 위상정렬은 “크기 기준 정렬”이 아니다.
**“앞에 와야 하는 것만 앞에 오면 된다”**

> 각 노드와 각 간선이 한 번씩만 처리된다 👉 O(V + E)




```python
# 위상 정렬
from collections import deque

node, edge  = map(int, input().split())
# 진입 차수
indegree = [0] * (node+1)
# graph[시작 노드] = 도착 노드 list
graph = [[] for i in range(node+1)]

for _ in range(edge):
    a, b = map(int, input().split())
    graph[a].append(b)
    indegree[b] += 1

def topology_sort():
    result = []
    q = deque()

    for i in range(1,node+1): # O(V)
        if indegree[i] == 0:
            q.append(i)

    while q:
        now = q.popleft()
        result.append(now)
        for i in graph[now]: # O(E)
            indegree[i] -= 1
            if indegree[i] == 0:
                q.append(i)

    for i in result:
        print(i, end = ' ')

topology_sort()
```

### 실전 문제 2. 팀결성

> 경로 압축 없을 때 find 연산의 시간 복잡도 : O(N) -> M번 연산 & 연산마다 2번의 find : O(N × M)

> 경로 압축 적용 후에는, O(M x a(N))으로 시간 복잡도가 감소
- 이때, 일반적으로 a(N) <= 5
- 따라서, 사실상 시간 복잡도가 거의 O(M)으로 감소

```python
N, M = map(int, input().split())
parent = [0] * (N+2)
for i in range(1, N+2):
    parent[i] = i

def find_parent(parent, x):
    if x != parent[x]:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union_0(parent, a, b):
    root_a = find_parent(parent, a)
    root_b = find_parent(parent, b)
    if root_a > root_b:
        parent[root_a] = root_b
    else:
        parent[root_b] = root_a

def check_same_1(parent, a, b):
    if find_parent(parent, a) == find_parent(parent, b):
        print("YES")
    else:
        print("NO")

for _ in range(M):
    op, a, b = map(int, input().split())
    if op == 0:
        union_0(parent, a, b)
    elif op == 1:
        check_same_1(parent, a, b)
```

### 실전문제 3. 도시 분할 계획
- 최소 비용으로 2개의 서로소 집합 그룹 만들기

> 🔥 **핵심 IDEA : “전체를 연결하되, 가장 비싼 간선 하나를 끊어라”**

```python
""" MST를 만든 후, 비용이 최대인 road 제거
그렇지 않고, 2개의 group을 따로 만들다보면, 예외가 존재 """
house, road = map(int, input().split())
road_list = []
for _ in range(road):
    a, b, cost = map(int, input().split())
    road_list.append((cost, a, b))

# 서로소 집합 구조
parent = [0] * (house+1)
for i in range(1, house+1):
    parent[i] = i

def find_parent(parent, x):
    if x != parent[x]:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    root_a = find_parent(parent, a)
    root_b = find_parent(parent, b)
    if root_a < root_b:
        parent[root_b] = root_a
    else:
        parent[root_a] = root_b

# 비용을 기준으로 오름차순 정렬
road_list.sort(key = lambda x : x[0])

total_cost = 0
max_cost = 0
# 모든 edge를 돌며 순회
for rd in road_list:
    cost, a, b = rd
    if find_parent(parent, a) != find_parent(parent, b):
        union(parent, a, b)
        total_cost += cost
        max_cost = max(max_cost, cost)
    else:
        continue
print(total_cost - max_cost)
```

### 실전 문제4. 커리큘럼

> - 입력으로 들어오는 값의 개수가 random 하다면?
> - arr = list(map(int, input().split()))

```python
from collections import deque
n = int(input())

lecture_time_list = [0] * (n+1) # 각 강의당 강의 시간
lecture_before_list = [[] for _ in range(n+1)] # 각 강의의 선수 과목
num_before_list = [0] * (n+1) # 각 강의의 남은 선수 과목
lecture_after_list = [[] for _ in range(n+1)] # 각 강의의 후속 과목
minimum_result_table = [0] * (n+1) # 각 강의의 최소 시간 result
num_before_list[0] = -1

# 선수 과목 입력 받기
for i in range(1, n+1):
    time, *before, end = map(int, input().split())
    before = list(before)
    lecture_time_list[i] += time
    lecture_before_list[i].extend(before)
    for bf in before:
        lecture_after_list[bf].append(i)
    num_before_list[i] += len(before)

# 선수과목 0개인 것부터 q에 넣기
q = deque()
for i in range(1, n+1):
    if num_before_list[i] == 0:
        q.append(i)
        num_before_list[i] = -1

# q가 빌 때까지 반복
while q:
    now = q.popleft()
    if len(lecture_before_list[now]) > 0:
        for_max = [minimum_result_table[idx] for idx in lecture_before_list[now]]
        minimum_result_table[now] += lecture_time_list[now] + max(for_max)
    else:
        minimum_result_table[now] += lecture_time_list[now]

    for next_lecture in lecture_after_list[now]:
        num_before_list[next_lecture] -= 1 # 이떄 바로 0인지 확인해서 queue에 넣어도 됐음

    for i in range(1, n+1):         # n번의 연산이 while q에서 n번 반복
        if num_before_list[i] == 0: # 시간 복잡도 : O(N^2)
            q.append(i)
            num_before_list[i] -= 1


for i in range(1, n+1):
    print(minimum_result_table[i])
```

> 시간 복잡도 개선 Code

- 기존 코드 : 매번 전체 탐색 → O(N²) & 나중에 max 계산
- 수정 코드 : 간선 따라 처리 → O(V + E) & 진행하면서 minimum_result_table 값 누적
- 🔥 진입차수 0을 찾지 말고, 만들어지는 순간 처리하기

```python
from collections import deque
n = int(input())

lecture_time_list = [0] * (n+1)
num_before_list = [0] * (n+1)
lecture_after_list = [[] for _ in range(n+1)]
minimum_result_table = [0] * (n+1)

# 입력
for i in range(1, n+1):
    data = list(map(int, input().split()))
    lecture_time_list[i] = data[0]

    for bf in data[1:-1]:
        lecture_after_list[bf].append(i)
        num_before_list[i] += 1

# queue 초기화
q = deque()
for i in range(1, n+1):
    if num_before_list[i] == 0:
        q.append(i)
        minimum_result_table[i] = lecture_time_list[i]

# 위상 정렬 + DP
while q:
    now = q.popleft()

    for next_lecture in lecture_after_list[now]:
        num_before_list[next_lecture] -= 1

        # 매번, list 전체에서 max값을 찾기보다, 매번 갱신 -> O(V+E)
        """기존 → 매번 전체 다시 계산    개선 → incremental update"""
        minimum_result_table[next_lecture] = max( minimum_result_table[next_lecture], minimum_result_table[now] + lecture_time_list[next_lecture] )

        # 진입차수 0 되는 순간 queue
        if num_before_list[next_lecture] == 0:
            q.append(next_lecture)

# 출력
for i in range(1, n+1):
    print(minimum_result_table[i])
```
