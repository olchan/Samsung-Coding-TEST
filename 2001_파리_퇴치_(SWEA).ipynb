# 2001. 파리 퇴치 (SWEA)

- https://www.youtube.com/watch?v=4pKromHmu5E&list=PLodgw23vNd_UBfz8vwuylEXCtSfDc005w&index=7

```python
test_case = int(input())
for tc in range(test_case):
    # n : map 크기  # m : 파리채 크기
    n, m = map(int, input().split())
    # map 입력 받기
    map_fly = [[] for _ in range(n)]
    for i in range(n):
        map_fly[i].extend(list(map(int, input().split())))

    # maximum 갱신하며 탐색
    maximum = 0
    for row in range(n-m+1):
        for col in range(n-m+1):
            sum_fly = 0
            for i in range(m):
                sum_fly += sum(map_fly[row + i][col:col+m])
            maximum = max(maximum, sum_fly)

    print(f"#{tc+1}", maximum)
```
