> ### **선택 정렬**

- 매 반복마다 남은 것들 중에서 가장 작은 것을 선택하여 가장 앞의 위치로 보내는 과정을 반복하는 정렬 알고리즘

- 시간 복잡도 : N-1 + N-2 + ...+ 2 -> N(N-1) /2의 연산 -> O(N^2)

```python
# 남은 data 중 가장 작은 data를 앞으로 보내는 과정을 N - 1 번 반복하면 정렬이 완료됨
array = [7,5,9,0,3,1,6,2,4,8]

for i in range(len(array)):
  min_index = i
  for j in range(i+1, len(array)):
    if array[min_index] > array[j]:
      min_index = j
  array[i], array[min_index] = array[min_index], array[i]
  # SWAP

print(array)
```

> ### **삽입 정렬**

- 특정 data가 이전까지는 이미 정렬되어 있다고 가정한 상황에서, 그 data를 적절한 위치에 삽입하는 알고리즘으로, 필요한 경우에만 위치를 바꾸기에 data가 거의 정렬되어 있을 때 효율적

- 특정한 data가 자기보다 작은 data를 만나면, 그 위치에 삽입된다. 그렇기에, 이미 정렬이 거의 되어 있는 상태라면 매우 빠르게 동작하는 정렬 알고리즘.

> 시간 복잡도 : O(N^2)

```python
# swap 기반 삽입정렬
array = [7,5,9,0,3,1,6,2,4,8]

for i in range(1,len(array)):
  for j in range(i, 0, -1): # (start, end, step) : j -> 1까지 감소
    if array[j] < array[j-1]:
      array[j], array[j-1] = array[j-1], array[j]
    else:
      break

print(array)
```

```python
# 실제 삽입
array = [7,5,9,0,3,1,6,2,4,8]

for i in range(1, len(array)):
    key = array[i] # 삽입할 값
    j = i - 1 # 삽입할 값 이전의 index

    while j >= 0 and array[j] > key: # 0까지의 index 중에서,
    # 삽입할 값보다 이전 값이 클 때까지는 한칸씩 밀기
        array[j+1] = array[j]
        j -= 1

    array[j+1] = key
```

> ### **퀵 정렬**

- "**기준 data인 pivot을 설정**하고 그 기준보다 큰 데이터와 작은 데이터의 위치를 바꾸어서, 큰 문제를 작은 문제로 분할."

- 퀵 정렬을 수행하기 전에 pivot을 어떻게 설정할 것인지를 미리 명시해야 한다.

- pivot의 왼쪽에는 pivot 보다 작은 data가 위치하고, pivot의 오른쪽에는 pivot보다 큰 data가 위치하도록 하여 큰 문제를 작은 문제로 Divide.

> - Tree를 한 번 나누는 데에 드는 비용(Partition) : left와 right 포인터가 서로를 향해 가면서 배열 전체를 한 번 스캔하기에 O(N)
> - divide 단계의 경우, 평균적으로는 O(log2N) 만큼 나누게 됨.
> - 따라서, 평균적으로는 O(N log N)의 시간 복잡도
> - 다만, 만약 pivot이 항상 가장 작은 or 큰 값이면 최악의 경우 시간복잡도 : O(N^2)



```python
""" quick 정렬과 병합 정렬 모두 O(NlogN)의 시간 복잡도를 가지지만,
퀵 정렬이 메모리 이동이 더 적어서 더 빠르게 동작한다.
반대로, 병합정렬은 계속해서 복사를 하기에 느리게 동작한다."""
```

```python
# left, right 포인터가 교차하는 순간은
# 왼쪽 영역에는 pivot보다 큰 값이 없음 & 오른쪽 영역에는 pivot보다 작은 값이 없음
# 그렇기에, 큰값 작은값의 위치가 정상화되었으니 pivot이 들어갈 수 있으므로
# 가장 왼쪽에 있던 piviot을 right과 자리를 바꿔준다.

# 최종 결과
# [ pivot보다 작은 값들 ] <right>  pivot <left> [ pivot보다 큰 값들 ]
```

```python
# quick sort
array = [7,5,9,0,3,1,6,2,4,8]

def quick_sort(array, start, end):
  if start >=  end: # 원소가 1개인 경우 종료
    return

  pivot = start
  left = start + 1
  right = end

  # pivot : 가장 왼쪽에 있는 숫자
  # left에서 시작해서는 pivot 보다 큰 숫자 찾기
  # right에서 시작해서는 pivot 보다 작은 숫자 찾기

  # 찾은 left와 right의 순서가 left < right면
  # 두 값을 swap
  # right 인덱스 < left 인덱스 이면,
  # pivot 하고, right값 (pivot 보다 작은 값)을 swap

  while left <= right:  # O(N)의 비용
    while left <= end and array[left] <= array[pivot]:
      left += 1
    while right > start and array[right] >= array[pivot]:
      right -= 1

    if left > right:
      array[right], array[pivot] = array[pivot], array[right]
    else:
      array[left], array[right] = array[right], array[left]

  quick_sort(array, start, right-1)
  quick_sort(array, right+1, end)

quick_sort(array, 0, len(array)-1)
print(array)
```

```python
# python의 장점을 살린 퀵 정렬 코드
array = [7,5,9,0,3,1,6,2,4,8]

def quick_sort(array):
  if len(array) <= 1:
    return array

  pivot = array[0]
  tail = array[1:]

  left_side = [x for x in tail if x <= pivot]
  right_side = [x for x in tail if x >= pivot]

  return quick_sort(left_side) + [pivot] + quick_sort(right_side)

array = quick_sort(array)
print(array)
```

> ## **병합 정렬**

시간 복잡도
- 쪼개는 과정 : O(log2 N)
- 각 단계마다 쪼개진 배열들의 총 길이는 N이고,
그걸 다 훑으니까 N 연산 × logN 단계 → N log N

> **최악의 경우에도 O(NlogN)의 시간 복잡도를 가짐**


```python
# 문제를 작은 문제로 분할하는 용도
def merge_sort(array):
    if len(array) <= 1:
        return array

    mid = len(array) // 2

    # 왼쪽 / 오른쪽 분할
    left = merge_sort(array[:mid])
    right = merge_sort(array[mid:])

    # 정렬하면서 merge
    return merge(left, right)

# 각 merge 단계에서 “정렬된 배열”을 반환하고,
# 그 결과가 상위 호출의 left/right로 들어가면서 전체가 정렬된다.

# 순서에 맞게 합치는 용도
def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 남은 원소 추가
    result += left[i:]
    result += right[j:]
    return result

array = [7,5,9,0,3,  1,6,2,4,8]

sorted_array = merge_sort(array)
print(sorted_array)
```

> ### **계수 정렬 Counting Sort** : **데이터가 정수이며, 값의 범위가 작고 제한되어 있을 때는 기본 정렬 라이브러리보다 더 빠르게 동작하기에 사용하는 정렬 알고리즘**

- data 개수가 N, data의 최대값이 K일 때, 최악의 경우이더라도 O(N+K)의 시간 복잡도를 보장하는 정렬 알고리즘

> - data의 모든 범위를 담을 수 있는 크기의 list를 선언한 후 그 안에 정렬에 대한 정보를 담아서 정렬을 하는 알고리즘으로 비교 기반의 정렬 알고리즘이 아니다.

```python
""" 계수 정렬 : data의 크기가 한정되어 있고, data 안의 중복이 많이 있을수록 유리 """
""" quick sort : data의 특성을 파악하기 어려울 때 사용하기 적절한 정렬 방법 """
```

```python
# 가장 큰 data와 가장 작은 data가 모두 담길 수 있도록 list를 생성
array = [7,5,9,0,3,1,6,2,9,1,4,8,0,5,2]
maximum = max(array)
minimum = min(array)
count = [0] * (maximum - minimum + 1)

for i in range(len(array)):
  count[array[i] - minimum] += 1

for i in range(len(count)):
  for joongbok in range(count[i]):
    print(i + minimum, end = " ")
```

### 실전 문제 2. 위에서 아래로

```python
n = int(input())
num_list = []
for i in range(n):
    num_list.append(int(input()))

minimum = min(num_list)

maximum = max(num_list)

sorting = [0 for _ in range(maximum - minimum + 1)]

for num in num_list:
    sorting[num - minimum] +=1

for i in range(len(sorting) -1 , -1, -1):
    for j in range(sorting[i]):
        print(i + minimum)
```
